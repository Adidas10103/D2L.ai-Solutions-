# #4

PyTorch.autograd
[https://pytorch.org/docs/stable/notes/autograd.html]

```python
import torch
```

```python
def loss(y_hat, y):
    l = (y_hat - y.reshape(y_hat.shape)) **2 /2
    return l.mean()

x = torch.tensor([[1.,2.], [2.,9.]])
b = torch.tensor([1.]).requires_grad_(True)
w = torch.tensor([1.,4.]).reshape(-1,1).requires_grad_(True)
y_hat = (torch.matmul(x, w) + b)
y = torch.tensor([3.4, 7.2])

loss = loss(y_hat, y)
loss.backward(retain_graph=True)
print(w.grad  ,"\n" , b.grad)
loss.backward()
print(w.grad,"\n" , b.grad)
```
